```python
import numpy as np
import matplotlib.pyplot as plt
```

### Step Function

```python
# not used in practice

def step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-4, 4, 100)
plt.plot(x,step(x))
plt.ylim([-1,2])
plt.grid(True, alpha=0.5, linestyle='--')
```

### Sigmoid Function

```python
# typically used in last layer of a binary classification model
# nn.Sigmoid() / input.sigmoid()

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-4, 4, 100)
plt.plot(x, sigmoid(x))
plt.ylim([-1,2])
plt.grid(True, alpha=0.5, linestyle='--')
```

### TanH Function

```python
# scaled sigmoid, typically used in hidden layers
# nn.TanH() / input.tanh()

def tanh(x):
    return 2 / (1 + np.exp(-2*x)) -1

x = np.linspace(-4, 4, 100)
plt.plot(x, tanh(x))
plt.ylim([-2,2])
plt.grid(True, alpha=0.5, linestyle='--')
```

### ReLU Function

```python
# most popular choice for hidden layers
# nn.ReLU / input.relu()

def relu(x):
    return np.maximum(x, 0)

x = np.linspace(-4, 4, 100)
plt.plot(x, relu(x))
plt.ylim([-2,5])
plt.grid(True, alpha=0.5, linestyle='--')
```

### Leaky ReLU Function

```python
# improved version of ReLU, tries to solve vanishing gradient problem
# nn.LeakyReLU / F.leaky_relu

def leaky_relu(x, a):
    return np.where(x >=0, x, a*x)

x = np.linspace(-4, 4, 100)
plt.plot(x, leaky_relu(x, 0.1))
plt.ylim([-2,5])
plt.grid(True, alpha=0.5, linestyle='--')
```
